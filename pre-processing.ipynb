```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('/Users/kavyapatel/Desktop/washu spring 25/data mining/project2/speed_dating_dataset.csv', encoding='latin1')

# Get numeric columns
numeric_cols = df.select_dtypes(include='number').columns
num_vars = len(numeric_cols)

# Define grid layout: 3 columns
cols = 3
rows = math.ceil(num_vars / cols)

# Set figure size (you can tweak these)
fig, axes = plt.subplots(rows, cols, figsize=(12, rows * 3))
axes = axes.flatten()  # flatten in case rows > 1

for i, column in enumerate(numeric_cols):
    ax = axes[i]
    df[column].hist(bins=30, edgecolor='black', ax=ax)
    ax.set_title(f'{column}')
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.grid(False)

# Hide unused subplots (if any)
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
